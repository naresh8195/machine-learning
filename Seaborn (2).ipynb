# Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## PairGrid

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)

g.map(plt.scatter, color = 'green')
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
```

```python
g = sns.PairGrid(iris)

g.map_upper(plt.scatter)
```

```python
g = sns.PairGrid(iris)

g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## Faset Grid

```python
sns.set(font_scale = 1)
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')

g.map(sns.distplot, 'total_bill')
```

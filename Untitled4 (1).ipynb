```python
x1 = np.linspace(-np.pi, np.pi, 143)
x2 = np.linspace(-np.pi, np.pi, 143)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.figure(figsize = (15, 8), dpi = 120)

plt.plot(x1, y1, color = 'purple', lw = 4.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 4.5, ls = '-.')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--', label = '$y = cosx$')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.', label = '$y = sinx$')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Graph')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])

plt.legend(loc = 0)

plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
r = np.linspace(0, 8, 30)
theta = np.linspace(-0.8*np.pi, 0.7*np.pi, 50)

r,theta = np.meshgrid(r, theta)

X = r*np.sin(theta)
Y = r*np.cos(theta)
Z = f(X,Y)
```

```python
fig = plt.figure(figsize = (10, 68))

ax = plt.axes(projection = '3d')

ax.plot_surface(X, Y, Z, cmap = 'grey')

plt.show()
```
